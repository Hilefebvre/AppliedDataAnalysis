```python
import requests
import re
import json
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

Get the url for json data from Chrome Inspector Tool

```python
URL = 'https://www.topuniversities.com/sites/default/files/qs-rankings-data/357051.txt'
```

Get request to get content of the webpage about top unviersities rankings

```python
r = requests.get(URL)
```

Convert json data to dict

```python
data = r.json()
```

Display data

```python
data
```

Create dataframe from dict and select only the top 200 rows

```python
df = pd.DataFrame.from_dict(data.get('data'))
df = df.iloc[0:200]
```

Display head of dataframe

```python
df.head()
```

Getting info on nb of faculty, nb of international faculty, nb of students and nb of international students. To do so, go over the url column and proceed to get requests to retrieve info on the numbers previously mentioned and then store them in separate lists

```python
nb_faculty=[]
nb_faculty_inter=[]
nb_students=[]
nb_students_inter=[]

for url in df['url']:
    p = requests.get('https://www.topuniversities.com'+url).text
    s = BeautifulSoup(p, 'html.parser')
    
    if(s.find("div", {"class": "total faculty"})!=None):
        nb_faculty.append(s.find("div", {"class": "total faculty"}).find("div",{"class": "number"}).text.strip().replace(',',''))
    else:
        nb_faculty.append('0')
    if(s.find("div", {"class": "inter faculty"})!=None):
        nb_faculty_inter.append(s.find("div", {"class": "inter faculty"}).find("div",{"class": "number"}).string.strip().replace(',',''))
    else:
        nb_faculty_inter.append('0')    
    if(s.find("div", {"class": "total student"})!=None):
        nb_students.append(s.find("div", {"class": "total student"}).find("div",{"class": "number"}).text.strip().replace(',',''))
    else:
        nb_students.append('0')
    if(s.find("div", {"class": "total inter"})!=None):
        nb_students_inter.append(s.find("div", {"class": "total inter"}).find("div",{"class": "number"}).string.strip().replace(',',''))
    else:
        nb_students_inter.append('0')
```

Add new columns to dataframe with lists created in the previous step in order to correlate these new info to the universities they belong to

```python
df['Nb Faculty']=nb_faculty
df['Nb Faculty inter']=nb_faculty_inter
df['Nb Students']=nb_students
df['Nb Students Inter']=nb_students_inter
```

```python
df.head()
```

Getting rid of columns that do not particularly interest us

```python
df.drop(['cc','core_id','guide','logo','nid','score','stars','url'], axis=1, inplace=True)
```

```python
df.head()
```

Convert nb columns to numeric type

```python
df['Nb Faculty'] = pd.to_numeric(df['Nb Faculty'])
df['Nb Faculty inter'] = pd.to_numeric(df['Nb Faculty inter'])
df['Nb Students'] = pd.to_numeric(df['Nb Students'])
df['Nb Students Inter'] = pd.to_numeric(df['Nb Students Inter'])
```

Get original dataframe stored in other dataframe for further operations

```python
df_original=df
```

Compute ratio: Faculty/Students and InterStudents/Students and store them in separate dataframe columns

```python
df['ratio Faculty/Students']=df['Nb Faculty']/df['Nb Students']
df['ratio Inter Students/Students']=df['Nb Students Inter']/df['Nb Students']
```

```python
df.head()
```

Sort dataframe by ratio Faculty/Students descending values

```python
df_sorted_a = df.sort_values(['ratio Faculty/Students'], ascending=0)
```

```python
df_sorted_a.head()
```

Sort dataframe by ratio InterStudents/Students descending values

```python
df_sorted_b = df.sort_values(['ratio Inter Students/Students'],ascending=0)
```

```python
df_sorted_b.head()
```

Group dataframe by country

```python
df_c_grouped = df_original.groupby(['country'])
```

```python
for country, group in df_c_grouped:
    print('country', country)
    print('group', group)
```

Compute ratio Faculty/Students wrt. countries

```python
ratio_faculty_students_c = df_c_grouped['Nb Faculty'].sum()/df_c_grouped['Nb Students'].sum()
```

Sort ratio Faculty/Students Series by descending values

```python
ratio_faculty_students_c.sort_values(ascending=False)
```

Plot Faculty/Students ratio in function of countries

```python
ax1 = ratio_faculty_students_c.plot(kind='bar', grid=True, figsize=(10, 6), title='Ratio: Faculty/Students')
ax1.set_ylabel("Ratio: Faculty/Students")
plt.show()
```

Compute ratio InterStudents/Students wrt. countries

```python
ratio_studentsInter_students_c = df_c_grouped['Nb Students Inter'].sum()/df_c_grouped['Nb Students'].sum()
```

Sort ratio InterStudents/Students Series by descending values

```python
ratio_studentsInter_students_c.sort_values(ascending=False)
```

Plot InterStudents/Students ratio in function of countries

```python
ax2 = ratio_studentsInter_students_c.plot(kind='bar', grid=True, figsize=(10, 6), title='Ratio: Students Inter/Students')
ax2.set_ylabel("Ratio: StudentsInter/Students")
plt.show()
```
